```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
from statsmodels.tsa.stattools import adfuller
import csv
import warnings; 
warnings.filterwarnings('ignore')

# Load the exchange_rate dataset
data = pd.read_csv("C:/Users/porte/Downloads/dataset/exchange_rate/exchange_rate.csv", index_col='date', parse_dates=True)

# Select the relevant columns for the model
columns = ['0', '1', '2', '3', '4', '5', '6', 'OT']
```

```python
def calculate_stats(dataframe):
    # Calculate mean, standard deviation, min, 25% quantile, median, 75% quantile, and max
    stats = dataframe.describe().transpose()
    
    # Calculate skewness and kurtosis
    skewness = dataframe.apply(skew)
    kurt = dataframe.apply(kurtosis)
    
    # Add skewness and kurtosis to the statistics
    stats['Skewness'] = skewness
    stats['Kurtosis'] = kurt
    
    # Calculate ADF test for each column
    adf_results = {}
    for col in dataframe.columns:
        result = adfuller(dataframe[col])
        adf_results[col] = result[0]
    
    # Add ADF test results to the statistics
    stats['ADF_Test'] = pd.Series(adf_results)
    
    return stats

# Assuming 'data' is your DataFrame
stats = calculate_stats(data[:int(len(data)*0.70)])

# Print the output
print('TRAIN:')
print(stats)

# Save to CSV
stats.to_csv('train_data_statistics.csv')

# Assuming 'data' is your DataFrame
stats = calculate_stats(data[int(len(data)*0.70):int(len(data)*0.80)+1])

# Print the output
print('VAL:')
print(stats)

# Save to CSV
stats.to_csv('val_data_statistics.csv')

# Assuming 'data' is your DataFrame
stats = calculate_stats(data[int(len(data)*0.80)+1:len(data)])

# Print the output
print('TEST:')
print(stats)

# Save to CSV
stats.to_csv('test_data_statistics.csv')
```
